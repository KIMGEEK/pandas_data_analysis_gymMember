```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('student_lifestyle_dataset.csv')

df.columns = ['Student_ID','Study_Hours_Per_Day','Extracurricular_Hours_Per_Day','Sleep_Hours_Per_Day','Social_Hours_Per_Day','Physical_Activity_Hours_Per_Day','GPA','Stress_Level']
print(df.head())

print('\n')
```

```python
# 데이터 자료형 확인
print(df.info())  
print('\n')

# 데이터 통계 요약정보 확인
print(df.describe())
print('\n')
```

```python
# Stress_Level 데이터의 자료형을 문자열에서 숫자로 변환(Low, Moderate, High -> 1, 2, 3)
df['Stress_Level'] = df['Stress_Level'].replace({
    'Low': 1,
    'Moderate': 2,
    'High': 3
})
print(df['Stress_Level'].dtype)
print('\n')

print(df.describe())
print('\n')
```

```python
ndf = df[['Study_Hours_Per_Day','Extracurricular_Hours_Per_Day','Sleep_Hours_Per_Day','Social_Hours_Per_Day','Physical_Activity_Hours_Per_Day','GPA','Stress_Level']]
print(ndf.head())   
print('\n')

ndf.plot(kind='scatter', x='GPA', y='Study_Hours_Per_Day', c='coral', s=10, figsize=(10, 5))
```

```python
ndf.plot(kind='scatter', x='GPA', y='Social_Hours_Per_Day', c='coral', s=10, figsize=(10, 5))
```

```python
sns.pairplot(ndf)
plt.show()
plt.close()
```
