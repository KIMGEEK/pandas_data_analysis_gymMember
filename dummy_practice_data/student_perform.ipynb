```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Cleaned_Students_Performance.csv')

df.columns = ['gender','race_ethnicity','parental_level_of_education','lunch','test_preparation_course','math_score',
              'reading_score','writing_score','total_score','average_score']
print(df.head())

print('\n')
```

```python
print(df.info())
print('\n')
print(df.describe())
print('\n')
```

```python
ndf = df[['parental_level_of_education', 'lunch', 'test_preparation_course', 'math_score', 'reading_score', 'writing_score', 'average_score']]
print(ndf.head())
print('\n')

df_parent_grouped = ndf.groupby(['parental_level_of_education'])
result = df_parent_grouped['math_score'].agg(['mean', 'max', 'min', 'count'])
print(result)

result.plot(kind='bar', y='mean', color='coral', figsize=(10, 10))

#ndf.plot(kind='bar', x='parental_level_of_education', y='writing_score',
#          color='coral', s=10, figsize=(10, 10))
plt.show()
plt.close()
```

```python
ndf.plot(kind='scatter', x='math_score', y='writing_score',
          color='coral', s=10, figsize=(10, 10))
```

```python
df_lunch_grouped = ndf.groupby(['lunch'])
result = df_lunch_grouped['math_score'].agg(['mean', 'max', 'min', 'count'])
print(result)

result.plot(kind='bar', y='mean', color='coral', figsize=(10, 10))

#ndf.plot(kind='bar', x='parental_level_of_education', y='writing_score',
#          color='coral', s=10, figsize=(10, 10))
plt.show()
plt.close()
```
